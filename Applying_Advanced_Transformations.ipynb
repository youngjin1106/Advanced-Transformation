```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
# load the hero data
hero = pd.read_csv('Data/superhero_info - superhero_info.csv')
hero.head()
```

```python
# load the power data
power = pd.read_csv('Data/superhero_powers - superhero_powers.csv')
power.head()
```

```python
# Split on the |
hero['Hero|Publisher'].str.split('|',expand = True)
```

```python
# Add split columns to hero df with appropriate names
hero[['Hero','Publisher']] = hero['Hero|Publisher'].str.split('|',expand = True)
hero.head(2)
```

```python
# drop the Hero|Publisher column from the original data
hero = hero.drop(columns = ['Hero|Publisher'])
hero.head(2)
```

```python
# Seperate Measurements into 2 columns 
hero.loc[0,'Measurements']
```

```python
# To use Json, I have to change the single quote to double quotes
hero['Measurements'] = hero['Measurements'].str.replace("'",'"')
hero['Measurements']
```

```python
# Convert it into a Python Dictionary 
hero['Measurements'] = hero['Measurements'].apply(json.loads)
hero.loc[0, 'Measurements']
```

```python
# unpack it into seperate columns
height_weight = hero['Measurements'].apply(pd.Series)
height_weight
```

```python
# add this to the original data
hero = pd.concat((hero,height_weight), axis = 1)
hero.head(2)
```

```python
# drop the Measurements column from the original data
hero = hero.drop(columns = ['Measurements'])
hero.head(2)
```

```python
# making height and weight as numeric
hero['Height'] = hero['Height'].str.replace('cm','')
hero['Height'].head(2)
```

```python
hero['Height'] = pd.to_numeric(hero['Height'])
```

```python
hero['Weight'] = hero['Weight'].str.replace('kg','')
hero['Weight'].head(2)
```

```python
hero['Weight'] = pd.to_numeric(hero['Weight'])
```

```python
hero.info()
```

```python
# one hot encode powers
power['Powers'].head(2)
```

```python
power.loc[0,'Powers']
```

```python
# split it on the ','
power['powers_split'] = power['Powers'].str.split(',', expand = False)
power['powers_split']
```

```python
# seperate row for each value in the list columns 
exploded = power.explode('powers_split')
exploded
```

```python
# saving unique values
cols_to_make = exploded['powers_split'].dropna().unique()
cols_to_make
```

```python
# create a new column for each of our unique values that are stored in cols_to_make.
for col in cols_to_make:
    power[col] = power['Powers'].str.contains(col)
power.head()
```

```python
# combine the hero and power dataframe into one
merged = pd.merge(hero, power, left_on = 'Hero', right_on = 'hero_names')
merged.head()
```

## 1.Compare the average weight of super powers who have Super Speed to those who do not.

```python
merged.groupby('Super Speed')['Weight'].mean()
```

```python
sns.barplot(data = merged, x = 'Super Speed', y = 'Weight');
```
